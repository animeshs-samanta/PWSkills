```python
# 1. Find the value of T(2) for the recurrence relation T(n)=3T(n-1)+12n. given that T(0)=5.

T(n) = 3T(n-1) + 12n

T(1) = 3T(1-1) + 12*1
     = 3T(0) + 12
     = 3*5 + 12
     = 15 + 12
     = 27
        
T(2) = 3T(2-1) + 12*2
     = 3T(2) + 24
     = 3*27 + 24
     = 81 + 24
     = 105 (Ans)
        
```

```python
# 2. Given a recurrence relation, solve it using the substitution method:
    ''' a. T(n)=T(n-1)+c '''
    
T(n) = T(n-1) + c

Now,
T(n-1) = T[(n-1)-1] + c
       = T(n-2) + c
and,
T(n-2) = T[(n-2)-1] + c
       = T(n-3) + c
So, 
T(n) = T(n-1) + c
     = T(n-2) + c + c
     = T(n-2) + 2c
     = T(n-3) + c + 2c
     = T(n-3) + 3c
        
if we go upto k times then
T(n) = T(n-k) + kc

Assume n-k = 0
so, n=k
T(n) = T(n-n) + nc
     = T(0) + nc
     = 1+ nc
     = n
    
Complexity = O(n)
```

```python
# 2. Given a recurrence relation, solve it using the substitution method:
    ''' b. T(n)=2T(n/2)+n '''

T(n) = 2T(n/2) + n

Now,
T(n/2) = 2T[(n/2)/2] + n/2
       = 2T(n/2^2) + n/2


and,
T(n/2^2) = 2T[(n/2^2)/2] + n/2^2
         = 2T(n/2^3) + n/2^2
    
So, 
T(n) = 2T(n/2) + n
     = 2* [2T(n/2^2) + n/2] +n
     = 2^2 T(n/2^2) + n + n
     = 2^2 [2T(n/2^3) + n/2^2] + 2n 
     = 2^3 T(n/2^3) + n + 2n
     = 2^3 T(n/2^3) + 3n
    
if we go upto k times then
T(n) = 2^k T(n/2^k) + kn

Assume n/2^k = 1
    => n = 2^k
    => log(n) = k log(2)
    => k = log(n)
    
T(n) = n*T(1) + kn
     = n + n*log(n)
     = n*log(n)
        
Complexity = O(n*log(n))
```

```python
# 2. Given a recurrence relation, solve it using the substitution method:
    ''' c. T(n)=2T(n/2)+c '''
    
T(n) = 2T(n/2) + c

Now,
T(n/2) = 2T[(n/2)/2] + c
       = 2T(n/2^2) + c


and,
T(n/2^2) = 2T[(n/2^2)/2] + c
         = 2T(n/2^3) + c
    
So, 
T(n) = 2T(n/2) + c
     = 2* [2T(n/2^2) + c] + c
     = 2^2 T(n/2^2) + 2c + c
     = 2^2 [2T(n/2^3) + c] + 2c + c 
     = 2^3 T(n/2^3) + 2^2c + 2c + c
     = 2^3 T(n/2^3) + 4c + 2c + c
    
if we go upto k times then
T(n) = 2^k T(n/2^k) + c[ 2^k-1 + 2^k-2 + ... ... + 2^2 + 2^1 + 1]


Assume n/2^k = 1
    => n = 2^k
    => log(n) = k log2(2)
    => k = log2(n)
    
T(n) = 2^log2(n) + c[1 + 2 + 3 + ......... + log2(n) times
     = n* log2(2) + [2^0 + 2^1 + 2^2 + —– + log2(n) times] –> Geometric Progression(G.P.) 
     = O(n) + c(n)
     = O(n)
        
Complexity = O(n)
```

```python
# 2. Given a recurrence relation, solve it using the substitution method:
    ''' d. T(n)=T(n/2)+c '''
    
T(n) = T(n/2) + c

Now,
T(n/2) = T[(n/2)/2] + c
       = T(n/2^2) + c


and,
T(n/2^2) = T[(n/2^2)/2] + c
         = T(n/2^3) + c
    
So, 
T(n) = T(n/2) + c
     = T(n/2^2) + c + c
     = T(n/2^2) + 2c
     = T(n/2^3) + c+ 2c 
     = T(n/2^3) + 3c

if we go upto k times then
T(n) = T(n/2^k) + kc

Assume n/2^k = 1
    => n = 2^k
    => log(n) = k log2(2)
    => k = log2(n)

So,
T(n) = T(1) + log(n)
     = log(n)
        
Complexity = O(log(n))
```

```python
# 3. Given a recurrence relation, solve it using the recursive tree approach:
    ''' a. T(n)=2T(n-1)+1 '''

    
T(n) = 2T(n-1) + 1

                       T(n) -------1
                    /        \     
                T(n-1)      T(n-1) ----- 2
               /     \      /     \    
             T(n-2) T(n-2) T(n-2)  T(n-2)  ---- 4
            
            :         :      :      : 
            :         :       :       :   
                
                      k                --------- 2^k
                    
Now, 

T(n) = 1 + 2 + 4 + 8 + ...... ...... + 2^k
     = (2^k+1  - 1) / (2-1)       Asssume n-k =0 ; so n=k
     = 2^n+1  - 1
     = 2^n
    
Complexity = O(2^n)
```
