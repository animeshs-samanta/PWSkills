```python
# 5. How can you access the methods and attributes of a parent class from a child class in Python? Give an example.
'''
In Python, you can access the methods and attributes of a parent (or base) class from a child (or derived) class using the super() function. 
The super() function returns a temporary object of the superclass, which allows you to call its methods and access its attributes.
'''
class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        print("Generic animal sound")

class Dog(Animal):
    def __init__(self, species, breed):
        # Calling the constructor of the parent class (Animal) using super()
        super().__init__(species)
        self.breed = breed

    def make_sound(self):
        # Calling the overridden method of the parent class (Animal) using super()
        super().make_sound()
        print("Dog barks")

    def display_info(self):
        print(f"This is a {self.species} of breed {self.breed}")


my_dog = Dog(species="Canine", breed="Golden Retriever")

my_dog.make_sound()    
print("Species:", my_dog.species)
my_dog.display_info() 
```

```python
# 18. What is the "diamond problem" in multiple inheritance, and how does Python address it?
'''
The "diamond problem" is a term used in the context of multiple inheritance in object-oriented programming. It occurs when a class inherits from two
classes that have a common ancestor. If both parent classes provide an implementation for a method, and the child class does not override that method,
it becomes ambiguous which method should be called.
'''
class A:
    def method(self):
        print("Method in class A")

class B(A):
    def method(self):
        print("Method in class B")

class C(A):
    def method(self):
        print("Method in class C")

class D(B, C):
    pass


obj_d = D()
obj_d.method()


'''
How Python Addresses the Diamond Problem:
Python addresses the diamond problem by using a method resolution order (MRO) to determine the order in which base classes are considered when looking
for a method or attribute. The C3 linearization algorithm is used to calculate the MRO.

In Python, the MRO is accessible via the __mro__ attribute or the mro() method. When a method is called on an instance, Python searches for that method
in the MRO from left to right.
'''
class A:
    def method(self):
        print("Method in class A")

class B(A):
    def method(self):
        print("Method in class B")

class C(A):
    def method(self):
        print("Method in class C")

class D(B, C):
    pass


obj_d = D()
obj_d.method()

# Print the method resolution order (MRO)
print(D.__mro__)
```

```python
# 6. Create a Python class for a bank account and demonstrate abstraction by hiding the account balance and providing methods to deposit and withdraw funds.

from abc import ABC, abstractmethod

class BankAccount(ABC):
    def __init__(self, account_holder, account_number):
        self.account_holder = account_holder
        self.account_number = account_number
        self._balance = 0  # Starting with a zero balance

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    def get_balance(self):
        return self._balance

class SavingsAccount(BankAccount):
    def __init__(self, account_holder, account_number, interest_rate=0.02):
        super().__init__(account_holder, account_number)
        self.interest_rate = interest_rate

    def deposit(self, amount):
        self._balance += amount

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
        else:
            print("Insufficient funds.")

    def add_interest(self):
        self._balance += self._balance * self.interest_rate


savings_account = SavingsAccount(account_holder="Animesh Samanta", account_number="123456789")
print(f"Initial Balance: ${savings_account.get_balance()}")

savings_account.deposit(1000)
print(f"Balance after deposit: ${savings_account.get_balance()}")

savings_account.withdraw(500)
print(f"Balance after withdrawal: ${savings_account.get_balance()}")

savings_account.add_interest()
print(f"Balance after adding interest: ${savings_account.get_balance()}")
```

```python
# 19. Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.

from abc import ABC, abstractmethod

class LibrarySystem(ABC):
    def __init__(self, name):
        self.name = name
        self.books = {}

    @abstractmethod
    def add_book(self, book_title, author):
        pass

    @abstractmethod
    def borrow_book(self, book_title):
        pass

    @abstractmethod
    def return_book(self, book_title):
        pass

class SimpleLibrary(LibrarySystem):
    def add_book(self, book_title, author):
        if book_title not in self.books:
            self.books[book_title] = {"author": author, "available": True}
            return f"Added '{book_title}' by {author} to the library."
        else:
            return f"'{book_title}' already exists in the library."

    def borrow_book(self, book_title):
        if book_title in self.books and self.books[book_title]["available"]:
            self.books[book_title]["available"] = False
            return f"Borrowed '{book_title}' from the library."
        elif book_title in self.books:
            return f"'{book_title}' is currently unavailable."
        else:
            return f"'{book_title}' does not exist in the library."

    def return_book(self, book_title):
        if book_title in self.books and not self.books[book_title]["available"]:
            self.books[book_title]["available"] = True
            return f"Returned '{book_title}' to the library."
        elif book_title in self.books:
            return f"'{book_title}' is already available in the library."
        else:
            return f"'{book_title}' does not exist in the library."


library = SimpleLibrary(name="My Library")

print(library.add_book("The Great Gatsby", "F. Scott Fitzgerald"))
print(library.add_book("To Kill a Mockingbird", "Harper Lee"))

print(library.borrow_book("The Great Gatsby"))
print(library.borrow_book("The Catcher in the Rye"))

print(library.return_book("The Great Gatsby"))
print(library.return_book("The Catcher in the Rye"))
```

```python
# 9. Describe the concept of "delegation" in composition and how it simplifies the design of complex systems.

'''
Delegation in the context of object-oriented programming and composition refers to the practice of one object handing over responsibility for a particular task to another object.
Instead of a class inheriting behavior from a superclass, it relies on the collaboration with other classes to achieve functionality. Delegation is a powerful concept that simplifies
the design of complex systems in several ways:

Separation of Concerns:
    Delegation promotes a clear separation of concerns by assigning specific responsibilities to different objects. Each object is responsible for a well-defined task, making the
    codebase more modular and easier to understand.

Code Reusability:
    Delegation facilitates code reuse by allowing objects to be reused in different contexts. Rather than inheriting behavior, a class can delegate tasks to other classes, promoting a 
    more flexible and reusable design.
    
Flexibility and Adaptability:
    Delegation allows for greater flexibility and adaptability in the system. Objects can be easily replaced or modified without affecting the rest of the system, as long as they adhere
    to the same interface.
    
Dynamic Behavior:
    The behavior of a class can be dynamically changed by switching the delegate at runtime. This dynamic nature allows for more adaptable and extensible systems without the need for
    extensive modifications to existing code.
    
Reduced Coupling:
    Delegation can help reduce tight coupling between classes. Instead of relying on a fixed hierarchy of classes, objects can interact through well-defined interfaces, minimizing
    dependencies and making the system more maintainable.

'''
```

```python
# 12. Create a Python class for a university course, using composition to represent students, instructors, and course materials.

class UniversityCourse:
    def __init__(self, name, code):
        self.name = name
        self.code = code
        self.students = []
        self.instructor = None
        self.course_materials = []

    def add_student(self, student):
        self.students.append(student)

    def set_instructor(self, instructor):
        self.instructor = instructor

    def add_course_material(self, material):
        self.course_materials.append(material)

    def __str__(self):
        return f"University Course: {self.name} ({self.code})"

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

    def __str__(self):
        return f"Student: {self.name} (ID: {self.student_id})"

class Instructor:
    def __init__(self, instructor_id, name):
        self.instructor_id = instructor_id
        self.name = name

    def __str__(self):
        return f"Instructor: {self.name} (ID: {self.instructor_id})"

class CourseMaterial:
    def __init__(self, title, content):
        self.title = title
        self.content = content

    def __str__(self):
        return f"Course Material: {self.title}"


course = UniversityCourse(name="Introduction to Computer Science", code="CS101")

student1 = Student(student_id="S001", name="Animesh")
student2 = Student(student_id="S002", name="Akash")

instructor = Instructor(instructor_id="I001", name="Dr. Avijit")

material1 = CourseMaterial(title="Week 1 Lecture Notes", content="Introduction to Python")
material2 = CourseMaterial(title="Assignment Guidelines", content="Data Structures and Algorithms")

course.add_student(student1)
course.add_student(student2)
course.set_instructor(instructor)
course.add_course_material(material1)
course.add_course_material(material2)

print(course)

# Print students, instructor, and course materials
print("Students:")
for student in course.students:
    print(f"  {student}")

print("Instructor:")
print(f"  {course.instructor}")

print("Course Materials:")
for material in course.course_materials:
    print(f"  {material}")
```

```python
# 14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.

class Restaurant:
    def __init__(self, name):
        self.name = name
        self.menus = []

    def add_menu(self, menu):
        self.menus.append(menu)

    def __str__(self):
        return f"Restaurant: {self.name}"

class Menu:
    def __init__(self, name):
        self.name = name
        self.dishes = []

    def add_dish(self, dish):
        self.dishes.append(dish)

    def __str__(self):
        return f"Menu: {self.name}"

class Dish:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.ingredients = []

    def add_ingredient(self, ingredient, quantity):
        self.ingredients.append({"ingredient": ingredient, "quantity": quantity})

    def __str__(self):
        return f"Dish: {self.name} | Price: ${self.price}"

class Ingredient:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Ingredient: {self.name}"


restaurant = Restaurant(name="Gourmet Delight")

menu1 = Menu(name="Breakfast Menu")
menu2 = Menu(name="Dinner Menu")

dish1 = Dish(name="Pancakes", price=8.99)
dish1.add_ingredient(Ingredient(name="Flour"), quantity=1)
dish1.add_ingredient(Ingredient(name="Eggs"), quantity=2)
dish1.add_ingredient(Ingredient(name="Milk"), quantity=1)

dish2 = Dish(name="Grilled Salmon", price=19.99)
dish2.add_ingredient(Ingredient(name="Salmon"), quantity=1)
dish2.add_ingredient(Ingredient(name="Lemon"), quantity=1)
dish2.add_ingredient(Ingredient(name="Herbs"), quantity=1)

menu1.add_dish(dish1)
menu2.add_dish(dish2)

restaurant.add_menu(menu1)
restaurant.add_menu(menu2)

print(restaurant)

# Print menus and dishes
for menu in restaurant.menus:
    print(menu)
    for dish in menu.dishes:
        print(f"  {dish}")
        for ingredient in dish.ingredients:
            print(f"    {ingredient['quantity']} {ingredient['ingredient']}")
```

```python
# 16. Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.

class Character:
    def __init__(self, name, health=100):
        self.name = name
        self.health = health
        self.weapon = None
        self.armor = None
        self.inventory = []

    def equip_weapon(self, weapon):
        self.weapon = weapon

    def equip_armor(self, armor):
        self.armor = armor

    def add_to_inventory(self, item):
        self.inventory.append(item)

    def attack(self, target):
        if self.weapon:
            damage = self.weapon.attack()
            target.take_damage(damage)
            return f"{self.name} attacks {target.name} with {self.weapon.name} and deals {damage} damage."
        else:
            return f"{self.name} attacks {target.name} with bare hands, but has no weapon."

    def take_damage(self, damage):
        if self.armor:
            reduced_damage = self.armor.reduce_damage(damage)
            self.health -= reduced_damage
            return f"{self.name} takes {reduced_damage} damage after armor reduction."
        else:
            self.health -= damage
            return f"{self.name} takes {damage} damage."

    def __str__(self):
        return f"Character: {self.name} | Health: {self.health}"

class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def attack(self):
        return self.damage

    def __str__(self):
        return f"Weapon: {self.name} | Damage: {self.damage}"

class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def reduce_damage(self, incoming_damage):
        return max(0, incoming_damage - self.defense)

    def __str__(self):
        return f"Armor: {self.name} | Defense: {self.defense}"


player = Character(name="Hero", health=150)
enemy = Character(name="Enemy", health=100)

sword = Weapon(name="Sword", damage=20)
shield = Armor(name="Shield", defense=10)

player.equip_weapon(sword)
player.equip_armor(shield)

print(player)
print(enemy)

print(player.attack(enemy))
print(enemy)

print(enemy.attack(player))
print(player)


potion = "Health Potion"
player.add_to_inventory(potion)
print(f"{player.name}'s Inventory: {player.inventory}")
```

```python
# 17. Describe the concept of "aggregation" in composition and how it differs from simple composition.

'''
Aggregation and simple composition are both concepts related to object-oriented programming, specifically in terms of how objects are related and combined. Let's delve into each concept:

Simple Composition:
    In simple composition, an object is composed of one or more other objects.
    The composed object has a "has-a" relationship with its parts. For example, a Car class may have instances of Engine, Wheel, and Chassis as its components.
    The lifecycle of the parts is tied to the lifecycle of the whole. If the Car is destroyed, its components are usually destroyed as well.
    Simple composition implies a strong relationship between the whole and its parts, and typically, the parts are considered integral to the whole.

    
Aggregation:
    Aggregation is a specific form of association, representing a relationship where one class is a part of another class but can exist independently.
    In aggregation, the "has-a" relationship still exists, but the parts can exist independently of the whole. For instance, consider a University class that has a collection of
    Department objects. A Department can exist even if the University is not present.
    
    The lifecycle of aggregated objects is not necessarily tied to the lifecycle of the whole. Destroying the University does not necessarily mean destroying all its Department objects.
    Aggregation implies a weaker relationship between the whole and its parts compared to simple composition
    
'''
```
