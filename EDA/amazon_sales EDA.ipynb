
3. What is the distribution of discounted prices vs. actual prices?

```python
df.columns
```

```python
df['discounted_price']
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','')
```

```python
#df['discounted_price'].astype(float)

df['discounted_price'] = pd.to_numeric(df['discounted_price'])
```

```python
df.discounted_price
```

```python
df.actual_price
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','')
```

```python
df['actual_price'] = pd.to_numeric(df['actual_price'])
```

```python
df['discounted_price'].describe()
```

```python
df['actual_price'].describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['actual_price'], kde = True, color='blue', label = "Actual Price", bins = 30)
sns.histplot(df['discounted_price'], kde=True, color='orange', label="Discounted Price", bins = 30)
plt.title('Distribution of Actual Prices vs. Discounted Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

4. How does the average discount percentage vary across categories?

```python
df.columns
```

```python
df['discount_percentage']
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','')
```

```python
df['discount_percentage'] =pd.to_numeric(df['discount_percentage'])
```

```python
df['discount_percentage']
```

```python
avg_discount_price_by_category = df.groupby('category')['discount_percentage'].mean().reset_index()
```

```python
avg_discount_price_by_category
```

```python
avg_discount_price_by_category
avg_discount_price_by_category = avg_discount_price_by_category.sort_values(by='discount_percentage', ascending=False)
```

```python
plt.figure(figsize=(100,40))
sns.barplot(x ='discount_percentage', y= 'category', data = avg_discount_price_by_category, palette='viridis', hue ='category', legend=False)

plt.title("Average Discount percentage by Category")
plt.xlabel("Average Discount Percentage")
plt.ylabel("Category")
plt.show()
```
